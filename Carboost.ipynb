## **Outliers handling**

```python
# This is especially common in skewed data like MentHlth and PhysHlth, where most people are at 0.
# Even a value like 10 days will keep being flagged as an “outlier” no matter how many you remove.
# Values near 30 are valid, but they may look like “outliers” statistically, because most people report 0 or just a few days.
# we use capping because Can't remove high values like 25 or 30 — they’re valid health reports.

#  independent copy

df = df.copy()


# Convert numeric columns to float

numeric_cols = ['BMI', 'MentHlth', 'PhysHlth']
df[numeric_cols] = df[numeric_cols].astype(float)


# Plot Before Outlier Handling

plt.figure(figsize=(12,6))
plt.suptitle("Before Outlier Handling", fontsize=16)
df [numeric_cols].boxplot()
plt.show()


# Outlier Handling (IQR Method)

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Cap outliers
    df.loc[:, col] = df[col].clip(lower, upper)


# Plot After Outlier Handling

plt.figure(figsize=(12,6))
plt.suptitle("After Outlier Handling", fontsize=16)
df[numeric_cols].boxplot()
plt.show()


```

```python
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import RandomizedSearchCV, cross_val_score, learning_curve
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# BASELINE MODEL

cat_base = CatBoostClassifier(
    iterations=200,       # number of trees
    learning_rate=0.1,
    depth=6,
    eval_metric='Accuracy',
    random_state=42,
    verbose=0
)

# Train baseline
cat_base.fit(X_train, y_train)

# Predict baseline
y_pred_cat_base = cat_base.predict(X_test)

# Evaluate
acc_cat_base = accuracy_score(y_test, y_pred_cat_base)
print(f"Baseline Test Accuracy: {acc_cat_base:.4f}")
print("\nClassification Report (Baseline):\n", classification_report(y_test, y_pred_cat_base))

# Confusion Matrix
cm_cat_base = confusion_matrix(y_test, y_pred_cat_base)
plt.figure(figsize=(5,4))
sns.heatmap(cm_cat_base, annot=True, fmt='d', cmap='Blues', xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Baseline CatBoost")
plt.show()
```

```python
# HYPERPARAMETER TUNING

from scipy.stats import randint, uniform

param_dist_cat = {
    'depth': randint(4, 10),
    'learning_rate': uniform(0.01, 0.2),
    'iterations': randint(200, 800),
    'l2_leaf_reg': uniform(1, 10)
}

cat_random = RandomizedSearchCV(
    CatBoostClassifier(
        eval_metric='Accuracy',
        random_state=42,
        verbose=0
    ),
    param_distributions=param_dist_cat,
    n_iter=20,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    random_state=42,
    verbose=1
)

cat_random.fit(X_train, y_train)
print("\nBest Parameters Found:", cat_random.best_params_)

# Best tuned model
best_cat = cat_random.best_estimator_
```

```python

#  EVALUATION OF TUNED MODEL

y_pred_cat_tuned = best_cat.predict(X_test)
acc_cat_tuned = accuracy_score(y_test, y_pred_cat_tuned)
print(f"\nTuned Test Accuracy: {acc_cat_tuned:.4f}")
print("\nClassification Report (Tuned):\n", classification_report(y_test, y_pred_cat_tuned))

cm_cat_tuned = confusion_matrix(y_test, y_pred_cat_tuned)
plt.figure(figsize=(5,4))
sns.heatmap(cm_cat_tuned, annot=True, fmt='d', cmap='Greens', xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Tuned CatBoost")
plt.show()
```

```python
#CROSS-VALIDATION RESULTS

cv_scores_cat = cross_val_score(best_cat, X_train, y_train, cv=5, scoring='accuracy', n_jobs=-1)
print("\nCross-validation scores:", np.round(cv_scores_cat, 4))
print(f"Mean CV Accuracy: {cv_scores_cat.mean():.4f} ± {cv_scores_cat.std():.4f}")
```

```python

# LEARNING CURVE

train_sizes, train_scores, test_scores = learning_curve(
    best_cat, X, y,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 5),
    shuffle=True,
    random_state=42
)

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.figure(figsize=(8,6))
plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Training Accuracy')
plt.plot(train_sizes, test_mean, 'o-', color='green', label='Validation Accuracy')

plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color='blue', alpha=0.1)
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color='green', alpha=0.1)

plt.title('Learning Curve for CatBoost Classifier')
plt.xlabel('Training Set Size')
plt.ylabel('Accuracy')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
